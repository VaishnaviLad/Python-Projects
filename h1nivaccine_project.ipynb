```python
X = vacc_df.drop(columns=['h1n1_vaccine'],axis=1)
y = vacc_df['h1n1_vaccine']
X_vacc, X_test, Y_vacc, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
module_0 = LogisticRegression()
```

```python
module_0.fit(X_vacc,Y_vacc)
```

```python
s0=module_0.score(X_vacc,Y_vacc)
print("After LogisticRegression vacc score are: ",s0)
```

```python
s0=module_0.score(X_test,y_test)
print("After LogisticRegression vacc score are: ",s0)
```

```python
Prediction = module_0.predict(X_test)
```

```python
accuracy = accuracy_score(y_test,Prediction)
```

```python
print(accuracy*100 , "% accuracy of the module")
```
