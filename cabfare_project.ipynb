we dont have null values but there are zeroes in it

```python
mean2 = cab_df["longitude_of_pickup"].mean()
cab_df["longitude_of_pickup"] = cab_df["longitude_of_pickup"].replace(np.nan,mean2)
```

```python
mean3 = cab_df["latitude_of_pickup"].mean()
cab_df["latitude_of_pickup"] = cab_df["latitude_of_pickup"].replace(np.nan,mean3)
```

```python
mean4 = cab_df["longitude_of_dropoff"].mean()
cab_df["longitude_of_dropoff"] = cab_df["longitude_of_dropoff"].replace(np.nan,mean4)
```

```python
mean5 = cab_df["latitude_of_dropoff"].mean()
cab_df["latitude_of_dropoff"] = cab_df["latitude_of_dropoff"].replace(np.nan,mean5)
```

```python
cab_df.head(15)
```

```python
 cab_df.drop(['date_time_of_pickup'],axis=1,inplace=True)
```

```python
cab_df.drop(['unique_id'],axis=1,inplace=True)
```

```python
cab_df.sample(10)
```

 implementation of the models

```python
x = cab_df.drop(columns=['amount'],axis=1)
y = cab_df['amount']
```

```python
X_cab, x_test, Y_cab,y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
module_1 = GradientBoostingRegressor()
```

```python
module_1.fit(X_cab,Y_cab)
```

```python
s1 = module_1.score(X_cab,Y_cab)
print("After GradientBoostingRegressor cab score are: ",s1)
```

```python
s1 = module_1.score(x_test,y_test)
print("After GradientBoostingRegressor cab score are: ",s1)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
module_2 = DecisionTreeRegressor()
```

```python
module_2.fit(X_cab,Y_cab)
```

```python
s2 = module_2.score(X_cab,Y_cab)
print("After Decision TreeRegressor cab score are: ",s2)
```

```python
s2 = module_2.score(x_test,y_test)
print("After Decision TreeRegressor cab score are: ",s2)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
module_3 = RandomForestRegressor()
```

```python
module_3.fit(X_cab,Y_cab)
```

```python
s3 = module_3.score(X_cab,Y_cab)
print("After RandomForestRegressor cab score are: ",s3)
```

```python
s3 = module_3.score(x_test,y_test)
print("After RandomForestRegressor cab score are: ",s3)
```
